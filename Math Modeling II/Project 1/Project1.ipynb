### Part A

```python
B = var('B')
F = var('F')
a = 10^-8
cB = 3000
cF = 15000
gB(B,F) = 0.05*B*((B - cB)/(B + cB))*(1 - (B/150000))
gF(B,F) = 0.08*F*((F - cF)/(F + cF))*(1 - (F/400000))
f1(B,F) = gB(B,F) - a * B * F
f2(B,F) = gF(B,F) - a * B * F
```
